Importing the Dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

Data Collection and Analysis

```python
#loading the data from csv file to a Pandas Dataframe
customer_data = pd.read_csv('/content/drive/MyDrive/Machine Learning Projects/Project 13 Customer Segmentation using K-means Clustering/Mall_Customers.csv')

#load the first 5 rows of the data
customer_data.head()
```

```python
# getting some informations about the dataset
customer_data.info()
```

```python
# checking for missing values
customer_data.isnull().sum()
```

Choosing the Annual Income Column and Spending Score Column

```python
#Group of customers will be classified based on their income and spending score
X = customer_data.iloc[:,[3,4]].values #taking the data and locating particular columns such as 3 and 4. indexing starts from 0 in python.
# the colon tells that we are taking columns. otherwise, rows will be returned.
```

```python
print(X)
```

Choosing the correct number of clusters

WCSS --> Within Clusters Sum of Squares

```python
# finding the wcss value for different number of clusters


wcss = []

# want the number of clusters to be 10
for i in range(1,11): #take all values form 1 to 10
  kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state=42)
  kmeans.fit(X)

  wcss.append(kmeans.inertia_) #gives us the wcss value; inertia lies in kmeans algorithm;
```

```python
#plot an elbow graph
sns.set()
plt.plot(range(1,11),wcss) #wcss is the y axis
plt.title('The Elbow Point Graph')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

Optimum Number of Clusters = 5

Training the K-means Clustering model ~ example of unsupervised learning model

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)

#return a label for each data point based on their cluster

Y = kmeans.fit_predict(X) #returns a cluster number for each of these datapoints
print(Y)
```

5 Clusters - 0,1,2,3,4

Visualizing all the Clusters

```python
# x axis is annual income and y axis is spending score

# plotting all the clusters and their centroids

plt.figure(figsize = (8,8))
plt.scatter(X[Y==0,0],X[Y==0,1],s=50,c='green', label = 'Cluster 1')
plt.scatter(X[Y==1,0],X[Y==1,1],s=50,c='red', label = 'Cluster 2')
plt.scatter(X[Y==2,0],X[Y==2,1],s=50,c='yellow', label = 'Cluster 3')
plt.scatter(X[Y==3,0],X[Y==3,1],s=50,c='violet', label = 'Cluster 4')
plt.scatter(X[Y==4,0],X[Y==4,1],s=50,c='blue', label = 'Cluster 5')

#plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='black',label = 'Centroids')
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
