Under-Sampling

Build a sample dataset containing similar distribution of normal transaction and Fraudulent transaction.

Number of Fradulent Transactions --> 492

```python
#randomly take 492 legit transactions for uniform distribution

legit_sample = legit.sample(n = 492) #randomly take 492 datapoints
```

Concatenating two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0) #axis = 0: add dataset one by one (row wise)
# if axis = 1, then the values will be added column wise.
new_dataset.head(10)
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean() #the nature of the dataset not changed compared to the one above.
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns = 'Class', axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_Test = train_test_split(X,Y, test_size=0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logisitc Regression

```python
model = LogisticRegression()
```

```python
#Training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('The accuracy score is ', training_data_accuracy) #out of 100, model can predict about 94 predictions
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_Test)

print('The test accuracy score is ', test_data_accuracy)

#why training accuracy? because if test and training data accuracy are too different, that means the training overtrained.
```
