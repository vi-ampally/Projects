Importing the Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data Collection and Processing

```python
# loading the csv data to a Pandas DataFrame
heart_data = pd.read_csv('/content/drive/MyDrive/Machine Learning Projects/Project 9 Heart Disease Prediction/heart_disease_data.csv')
```

```python
#print first 5 rows of the dataset
heart_data.head()
```

```python
#number of rows and columns in the dataset
heart_data.shape
```

```python
#check for null values
heart_data.isnull().sum()
```

```python
# getting some info about the data
heart_data.info()
```

```python
# statistical measures about the data
heart_data.describe()
```

```python
#checking the distribution of target variable
heart_data.target.value_counts()

#the distribution is almost equal
```

1 --> Defective Heart

0 --> Healthy Heart

Splitting the features and target

```python
heart_data.head()
```

```python
X = heart_data.drop(columns = 'target', axis = 1)
Y = heart_data['target']
```

```python
print(X)
print(Y)
```

Splitting the data into Training and Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.2, stratify = Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training


Logistic Regression Model

```python
model = LogisticRegression()
```

```python
#training the logisitcs regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
#Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
training_data_accuracy #model can predict values for about 85% of the data
```

```python
print('Accuracy on training data:', training_data_accuracy)
```

```python
#Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print('Accuracy on test data:', test_data_accuracy)
# almost 82%
#accuracy score for training and test should be almost same. if they are far away, that means the model overfitted.
```

Making a Predictive System

```python
input_data = (56,1,3,120,193,0,0,162,0,1.9,1,0,3) #reshape it because it is in a tuple

# changing the input_data to a numpy array as it is faster and easy
input_data_as_numpy_array = np.asarray(input_data)

#reshape the np array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
#print(prediction)

if(prediction == 0):
  print('The person does not have Heart Diseae.')
else:
  print('The person has Heart Disease.')
```
